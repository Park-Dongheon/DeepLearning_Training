```python
import pandas as pd
```

```python
data = pd.read_csv('data/GOOGL.csv')
data["Date"] = data.index
data.reset_index(drop=True, inplace=True)
data.tail()
```

```python
data = data[["Date", "Close"]]
data.info()
```

```python
import matplotlib.pyplot as plt
plt.plot(data["Date"], data["Close"])
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(data["Close"], model="multiplicative", period=30)
result.plot()
```

```python
pd.plotting.autocorrelation_plot(data["Close"])
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(data["Close"], lags=70)
```

```python
p, d, q = 5, 1, 2

from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(data["Close"], order=(p, d, q))
fitted = model.fit()
predicts = fitted.predict(len(data), len(data) + 10)
predicts
```

```python
data["Close"].plot(legend=True, label="Train")
predicts.plot(legend=True, label="Train")
```

```python

import statsmodels.api as sm

model = sm.tsa.statespace.SARIMAX(
    data["Close"], order=(p, d, q), seasonal_order=(p, d, q, 12)
)
fitted = model.fit()
predicts = fitted.predict(len(data), len(data) + 10)
data["Close"].plot(legend=True, label="Train")
predicts.plot(legend=True, label="Train")
```
